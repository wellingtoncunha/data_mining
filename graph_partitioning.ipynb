```python
from itertools import combinations
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
# Create Nodes and Edges 
nodes = ["a", "b", "c", "d", "e", "f", "g", "h"]
edges = [
    ["a", "b"],
    ["a", "c"],
    ["a", "d"],
    ["a", "e"],
    ["b", "c"],
    ["b", "d"],
    ["c", "d"],
    ["c", "h"],
    ["d", "g"],
    ["e", "f"],
    ["e", "g"],
    ["f", "g"],
]
edges = pd.DataFrame(edges, columns = ["to", "from"])
```

```python
df = pd.DataFrame(columns = ["from", "to", "cut_size"])
for i in range(1, len(nodes)):
    for each_from_node in list(combinations(nodes, i)):
        to_nodes = [i for i in nodes if i not in each_from_node]
        
        # If there is a record with a value from-to skip the calculation for the other way round to-from
        if len((df[(df["from"] == str(to_nodes)) & (df["to"] == str(list(each_from_node)))]).index) > 0:
            continue  
            
        # Get the cut_size it doesn't matter the direction
        cut_size = len(
            (edges[((edges["to"].isin(to_nodes)) & (edges["from"].isin(each_from_node))) | 
                  ((edges["to"].isin(each_from_node)) & (edges["from"].isin(to_nodes)))]).index
        )
    
        # Get the number of edges for the first (from) cluster
        from_edges = len(
            (edges[((edges["to"].isin(each_from_node)) & (edges["from"].isin(each_from_node))) | 
                  ((edges["to"].isin(each_from_node)) & (edges["from"].isin(each_from_node)))]).index
        )
        
        # Get the number of edges for the second (to) cluster
        to_edges = len(
            (edges[((edges["to"].isin(to_nodes)) & (edges["from"].isin(to_nodes))) | 
                  ((edges["to"].isin(to_nodes)) & (edges["from"].isin(to_nodes)))]).index
        )        
        
#         if cut_size > 0:
        df = df.append({
                "from": str(list(each_from_node)),
                "to": str(to_nodes),
                "cut_size": cut_size,
                "from_edges": 1 if from_edges == 0 else from_edges,
                "to_edges": 1 if to_edges == 0 else to_edges
            }, ignore_index = True
        )
        cut_size = 0
```

```python
# Calculate Sparsity
df["sparsity"] = df["cut_size"] / df[["from_edges", "to_edges"]].min(axis=1)
df.sort_values(["sparsity"])
```
